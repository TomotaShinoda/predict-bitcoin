```python
from assets.database import db_session
from assets.models import Data
#from keras.models import Sequential
```

```python
data = db_session.query(Data.date, Data.real_price, Data.pred_price).all()
```

```python
data
```

```python
dates = []
real_prices = []
pred_prices = []

for _data in data:
    dates.append(_data.date)
    real_prices.append(_data.real_price)
    pred_prices.append(_data.pred_price)
```

```python
print(real_prices[-2]) # 実際の今日の価格
```

```python
print(pred_prices[-1]) # 予測した明日の価格
```

```python
import datetime
DIFF_JST_FROM_UTC = 9
jst_date = dates[-1] + datetime.timedelta(hours=DIFF_JST_FROM_UTC)
print(jst_date)
print(jst_date.strftime("%Y/%m/%d %H:%M:%S"))
```
